```python
# creation
spark.sql("CREATE SCHEMA IF NOT EXISTS workspace.pace_and_space")
spark.sql("SHOW SCHEMAS IN workspace").show()
```

```python
# three medallion layer schemas
# Convention: catalog.layer_projectname
spark.sql("CREATE SCHEMA IF NOT EXISTS workspace.bronze_pace_and_space")
spark.sql("CREATE SCHEMA IF NOT EXISTS workspace.silver_pace_and_space")
spark.sql("CREATE SCHEMA IF NOT EXISTS workspace.gold_pace_and_space")

# Clean up the test schema we created a moment ago
spark.sql("DROP SCHEMA IF EXISTS workspace.pace_and_space")

spark.sql("SHOW SCHEMAS IN workspace").show()
```

```python
spark.sql("DROP SCHEMA IF EXISTS workspace.bronze_pace_and_space")
spark.sql("DROP SCHEMA IF EXISTS workspace.silver_pace_and_space")
spark.sql("DROP SCHEMA IF EXISTS workspace.gold_pace_and_space")
spark.sql("DROP SCHEMA IF EXISTS workspace.pace_and_space")

spark.sql("CREATE SCHEMA IF NOT EXISTS workspace.bronze_nba")
spark.sql("CREATE SCHEMA IF NOT EXISTS workspace.silver_nba")
spark.sql("CREATE SCHEMA IF NOT EXISTS workspace.gold_nba")

spark.sql("SHOW SCHEMAS IN workspace").show()
```

```python
spark.createDataFrame(
    [("environment_test", 1)], 
    ["name", "value"]
).write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.bronze_nba.environment_test")

print("Delta write to Unity Catalog successful")

spark.table("workspace.bronze_nba.environment_test").show()
```
